

### **1. What is a random variable in probability theory?**

A random variable is a variable that takes on numerical values determined by the outcome of a random phenomenon. It maps outcomes from a sample space to real numbers.

---

### **2. What are the types of random variables?**

There are two main types:

1. **Discrete random variables** – take on countable values (e.g., number of heads in coin tosses).
2. **Continuous random variables** – take on any value within a given range (e.g., height, weight).

---

### **3. What is the difference between discrete and continuous distributions?**

* **Discrete distributions** describe probabilities for variables that can take on specific, countable values.
* **Continuous distributions** describe probabilities for variables that can take on any value in a continuum or interval.

---

### **4. What are probability distribution functions (PDF)?**

A probability distribution function (PDF) describes the likelihood of a random variable taking a particular value.

* For discrete variables, it's called a **probability mass function (PMF)**.
* For continuous variables, it's called a **probability density function (PDF)**.

---

### **5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

* **PDF** shows the probability of a value exactly occurring.
* **CDF** gives the probability that a value will be less than or equal to a particular number.

---

### **6. What is a discrete uniform distribution?**

A discrete uniform distribution assigns equal probability to all outcomes in a finite set. For example, the roll of a fair die has a uniform distribution over {1, 2, 3, 4, 5, 6}.

---

### **7. What are the key properties of a Bernoulli distribution?**

* Only two possible outcomes: success (1) or failure (0).
* Single trial with success probability **p** and failure probability **1 – p**.
* Mean = p, Variance = p(1 – p).

---

### **8. What is the binomial distribution, and how is it used in probability?**

The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials. It's used when each trial has the same probability of success.

---

### **9. What is the Poisson distribution and where is it applied?**

The Poisson distribution models the number of events occurring in a fixed interval of time or space, given a constant mean rate and independence of events. It's often used for rare events (e.g., number of emails per hour).

---

### **10. What is a continuous uniform distribution?**

A continuous uniform distribution assigns equal probability to all values within a specific interval \[a, b]. The PDF is constant across the range and zero elsewhere.

---

### **11. What are the characteristics of a normal distribution?**

* Bell-shaped, symmetric curve
* Mean = median = mode
* Defined by its mean (μ) and standard deviation (σ)
* Total area under the curve equals 1

---

### **12. What is the standard normal distribution, and why is it important?**

The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It's important because it allows the use of Z-scores and simplifies statistical calculations.

---

### **13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size grows, regardless of the original distribution. It's vital for hypothesis testing and confidence intervals.

---

### **14. How does the Central Limit Theorem relate to the normal distribution?**

The CLT explains why many statistical methods assume normality: even if data is not normal, the sampling distribution of the mean becomes approximately normal for large sample sizes.

---

### **15. What is the application of Z statistics in hypothesis testing?**

Z-statistics are used to determine how far a sample mean is from the population mean in terms of standard deviations. They help calculate p-values and test hypotheses under normality assumptions.

---

### **16. How do you calculate a Z-score, and what does it represent?**

**Z = (X – μ) / σ**
It tells you how many standard deviations a value (X) is from the mean (μ). A Z-score standardizes data points across different distributions.

---

### **17. What are point estimates and interval estimates in statistics?**

* **Point estimate** gives a single best guess (e.g., sample mean) of a parameter.
* **Interval estimate** provides a range (confidence interval) within which the parameter is expected to lie.

---

### **18. What is the significance of confidence intervals in statistical analysis?**

Confidence intervals express the reliability of an estimate. For example, a 95% confidence interval means we are 95% confident that the interval contains the true parameter value.

---

### **19. What is the relationship between a Z-score and a confidence interval?**

Z-scores are used to determine the margins for confidence intervals in normal distributions. The Z-value defines how far the boundaries are from the mean.

---

### **20. How are Z-scores used to compare different distributions?**

By converting raw scores from different distributions to Z-scores, we can compare them on a common scale to understand relative performance or rarity.

---

### **21. What are the assumptions for applying the Central Limit Theorem?**

* Samples are independent
* Identically distributed
* Sample size is sufficiently large
* Finite mean and variance

---

### **22. What is the concept of expected value in a probability distribution?**

Expected value is the long-term average outcome of a random variable. It's calculated by summing all possible values weighted by their probabilities.

---

### **23. How does a probability distribution relate to the expected outcome of a random variable?**

The probability distribution defines the likelihood of each possible outcome, and the expected outcome (expected value) is the weighted average of those outcomes.

```python
#1. Write a Python program to generate a random variable and display its value
import numpy as np

random_variable = np.random.rand()
print("Random variable:", random_variable)
```

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import matplotlib.pyplot as plt

values = np.arange(1, 7)
pmf = np.ones_like(values) / len(values)

plt.stem(values, pmf, basefmt=" ")
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.grid(True)
plt.show()

```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

print("Bernoulli PDF (p=0.6, x=1):", bernoulli_pdf(0.6, 1))

```

```python
#4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram
from scipy.stats import binom

n, p = 10, 0.5
binom_data = binom.rvs(n, p, size=1000)

plt.hist(binom_data, bins=np.arange(n+2)-0.5, edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Successes")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
#5. Create a Poisson distribution and visualize it using Python
from scipy.stats import poisson

poisson_data = poisson.rvs(mu=3, size=1000)

plt.hist(poisson_data, bins=np.arange(poisson_data.min(), poisson_data.max()+2)-0.5, edgecolor='black', density=True)
plt.title("Poisson Distribution (mu=3)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
cdf = np.cumsum(pmf)

plt.step(values, cdf, where='post')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()

```

```python
#7. Generate a continuous uniform distribution using NumPy and visualize it
from scipy.stats import uniform

continuous_data = uniform.rvs(loc=0, scale=1, size=1000)

plt.hist(continuous_data, bins=30, edgecolor='black', density=True)
plt.title("Continuous Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()

```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
from scipy.stats import norm

x_val = 85
mean_val = 70
std_val = 10

z = (x_val - mean_val) / std_val
print("Z-score:", z)
print("Standard Normal CDF:", norm.cdf(z))

```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset
sample_mean = 72
population_mean = 70
std_dev = 10
n = 30

z_stat = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 1 - norm.cdf(z_stat)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result
data = np.random.normal(loc=100, scale=15, size=50)
mean = np.mean(data)
sem = np.std(data, ddof=1) / np.sqrt(len(data))
ci = (mean - 1.96 * sem, mean + 1.96 * sem)

print("95% Confidence Interval:", ci)

```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
normal_sample = np.random.normal(50, 8, 100)
mean = np.mean(normal_sample)
sem = np.std(normal_sample, ddof=1) / np.sqrt(100)
ci = (mean - 1.96 * sem, mean + 1.96 * sem)
print("Confidence Interval for Mean:", ci)
```

```python
# 21. Write a Python program to compare the standard deviations of two datasets and visualize the difference
a = np.random.normal(50, 5, 1000)
b = np.random.normal(50, 10, 1000)

plt.hist(a, alpha=0.5, label='std=5')
plt.hist(b, alpha=0.5, label='std=10')
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.grid(True)
plt.show()
```

```python
#22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
data = np.random.normal(0, 1, 1000)
data_range = np.ptp(data)
q75, q25 = np.percentile(data, [75, 25])
iqr = q75 - q25

print("Range:", data_range)
print("Interquartile Range (IQR):", iqr)

```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation
original = np.random.normal(100, 20, 1000)
normalized = (original - np.mean(original)) / np.std(original)

plt.hist(normalized, bins=30, edgecolor='black')
plt.title("Z-score Normalized Data")
plt.grid(True)
plt.show()
```
